```python
# Importamos nuevamente la librería
from nltk.stem import WordNetLemmatizer

# Descargar los recursos necesarios de NLTK (solo la primera vez)
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# Iniciamos el lematizador
lemmatizer = WordNetLemmatizer()

# Función para lematizar el texto
def lemmatize_text(text):
    # Tokenizar el texto
    tokens = nltk.word_tokenize(text)
    # Lematizar cada token
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)

# Aplicar la lematización a la columna 'contenido'
df_final['contenido_lemmatizado'] = df_final['contenido'].apply(lemmatize_text)  # revisar

# Mostrar el DataFrame con la columna lematizada
print(df_final)
```

```python
# Importamos librería
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Descargar los recursos necesarios de NLTK (solo la primera vez)
nltk.download('stopwords')
```

```python
# Obtener la lista de stop words en español
stop_words = set(stopwords.words('spanish'))

# Función para eliminar stop words
def remove_stopwords(text):
    tokens = word_tokenize(text)  # Tokenizar el texto
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

# Aplicar la eliminación de stop words a la columna 'contenido'
df_final['contenido_no_stopwords'] = df_final['contenido'].apply(remove_stopwords)

# Mostrar el DataFrame con la columna sin stop words
print(df_final)
```

```python
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Obtener la lista de stop words en español
stop_words = set(stopwords.words('spanish'))

# Función para normalizar el texto
def normalize_text(text):
    # Convertir a minúsculas
    text = text.lower()
    # Eliminar puntuación
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenizar el texto
    tokens = word_tokenize(text)
    # Eliminar stop words
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # Unir los tokens de nuevo en una cadena
    return ' '.join(filtered_tokens)

# Aplicar la normalización a la columna 'contenido'
df_final['contenido_normalizado'] = df_final['contenido'].apply(normalize_text)

# Mostrar el DataFrame con la columna normalizada
print(df_final)
```

```python
# Importamos la librería 
from collections import Counter
import matplotlib.pyplot as plt
from nltk.util import ngrams
```

```python
# Contar palabras frecuentes
all_words = ' '.join(df_final['contenido_normalizado']).split()
word_counts = Counter(all_words)
print(word_counts)
```

```python
# Extraer hashtags
hashtags = [word for text in df_final['contenido'] for word in text.split() if word.startswith('#')]
hashtag_counts = Counter(hashtags)

print(hashtag_counts)
```

```python
# Generar n-gramas
n = 2  # Cambia n para obtener diferentes n-gramas
all_ngrams = ngrams(all_words, n)
ngram_counts = Counter(all_ngrams)

print(ngram_counts)
```

```python
# Analizar longitud de documentos
df_final['longitud'] = df_final['contenido'].apply(len)

# Visualización de palabras frecuentes
plt.figure(figsize=(10, 5))
plt.bar(*zip(*word_counts.most_common(10)))
plt.title('Palabras Frecuentes')
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# Visualización de hashtags
plt.figure(figsize=(10, 5))
plt.bar(*zip(*hashtag_counts.most_common(10)))
plt.title('Hashtags Frecuentes')
plt.xlabel('Hashtags')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# Visualización de longitud de documentos
plt.figure(figsize=(10, 5))
plt.hist(df_final['longitud'], bins=5, alpha=0.7)
plt.title('Distribución de Longitud de Documentos')
plt.xlabel('Longitud')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Importamos la librería
import matplotlib.pyplot as plt
from collections import Counter
from nltk.sentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import nltk

import nltk
nltk.download('punkt')
nltk.download('vader_lexicon')
```

```python
# Convertir la columna 'fecha' a tipo datetime
df_final['fecha'] = pd.to_datetime(df_final['fecha'])
```

```python
# Inicializar el analizador de sentimientos
analyzer = SentimentIntensityAnalyzer()

# Función para obtener el sentimiento
def obtener_sentimiento(texto):
    puntuaciones = analyzer.polarity_scores(texto)
    return puntuaciones['compound']

# Aplicar la función de sentimiento al DataFrame
df_final['sentimiento'] = df_final['contenido'].apply(obtener_sentimiento)

# Clasificar el sentimiento
df_final['clasificacion_sentimiento'] = df_final['sentimiento'].apply(lambda x: 'positivo' if x > 0 else 'neutral' if x == 0 else 'negativo')

print(df_final)
```

```python
# Visualizar la distribución de sentimientos
plt.figure(figsize=(10, 5))
df_final['clasificacion_sentimiento'].value_counts().plot(kind='bar', color=['green', 'gray', 'red'])
plt.title('Distribución de Sentimientos')
plt.xlabel('Clasificación de Sentimiento')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.show()

# Generar una nube de palabras
text = ' '.join(df_final['contenido'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Mostrar la nube de palabras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de Palabras')
plt.show()

# Análisis de tendencias a lo largo del tiempo
df_final.set_index('fecha', inplace=True)
df_final.resample('D').mean()['sentimiento'].plot(figsize=(10, 5), marker='o')
plt.title('Tendencia del Sentimiento a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Sentimiento Promedio')
plt.grid()
plt.show()
```
